# 3. Extract the data from jan 2019 to apr 2021 and plot it. 

```python
e1=axis[(axis.year>2018)]
```

```python
e1.head()
```

```python
plt.figure(figsize=(20,5))
plt.plot(e1.Date,e1.Open,'r');
plt.xlabel("Date",fontdict = {"size":20});
plt.ylabel("Openibg ammount" ,fontdict = {"size":20});
```

```python
plt.figure(figsize=(20,5))
plt.plot(e1.Date,e1.Volume);
plt.xlabel("Date",fontdict = {"size":20});
plt.ylabel("Volume" ,fontdict = {"size":20});
```

```python
plt.figure(figsize=(20,5))
plt.plot(e1.Date,e1.Trades,"g");
plt.xlabel("Date",fontdict = {"size":20});
plt.ylabel("Trades" ,fontdict = {"size":20});
```

 ###   In abov graph we easly can see that Trades and Volume show a large increment change and the value of opening ammount  start drop in the start of 2020 due to the pandemic covid-19.

# 4. Plot the open ammount with Date field after grouping month and yearly .

```python
axis.head()
```

```python
plt.figure(figsize=(20,5))
plt.scatter(axis.year,axis.Open)
plt.xticks(range(2000,2022));
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.xlabel("year",fontdict = {"size":20})
plt.ylabel("Opening ammount" ,fontdict = {"size":20})
```

###  In scatter graph we see the opening ammount is very large in 2010 to 2013 as compear to 2000.But the opening ammount in the staring 2014 was decrease in large amount. 

```python

plt.figure(figsize=(20,5))
plt.scatter(axis.month,axis.Open)
plt.xticks(range(0,13));
plt.grid(color = 'red', linestyle = '--', linewidth = 0.5)
plt.xlabel("Month",fontdict = {"size":20})
plt.ylabel("Opening ammount" ,fontdict = {"size":20})
```

### BY seeing this scatter graph we can easly says month 5,6,7 is opening ammount is high as commper to other months.
