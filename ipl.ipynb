```python
#How many times chasing team has won the match?
count1=0
loss=0
x=0
y=0
total=ipl_bat.id.nunique()
for i in range(total):
    
    if ipl_bat.toss_decision[i]=="field":
        if ipl_bat.toss_winner[i]==ipl_bat.winner[i]:
            count1+=1
        else:
            loss+=1
    else :
        if ipl_bat.toss_winner[i]==ipl_bat.winner[i]:
            x+=1
        else:
            y+=1
print("Number of team which chasing and  win both toss and match:",count1)
print("Number of team which chasing and  loss toss but win match:",loss)
print("Number of team which chose bating and win both toss and match:",x)
print("Number of team which chose bating and loss toss but win match:",y)
```

```python
plt.figure(figsize=(7,7))
li=[count1,loss,x,y]
myexplode = [0.01,0.01,0.01,0.01]
plt.pie(li, labels=[f"chasing and win both match and toss {count1}",
                    f"chasing and win match but loss toss {loss}",
                    f"batting and win both match and toss {x}",
                    f"batting and win match but loss toss {y}"],
                   explode = myexplode,shadow=True);
```

```python
#Which team has played the most number of matches?
team =ipl_bat.team1.unique()
max=0
t_name=""
for i in team:
    t1=ipl_bat.groupby("team1").get_group(i).count().id
    t2=ipl_bat.groupby("team2").get_group(i).count().id
    T=t1+t2
    print (f"{i}:",T)
    if T>max:
        max=T
        t_name=i
print(f"\nTeam has played the most number of matches is {t_name} ",max)
ipl_bat.team1    
```

```python
#Which team has the highest winning percentage?
team =ipl_bat.team1.unique()
t_name=""
for i in team:
    w=ipl_bat.groupby("winner").get_group(i).count().id
    t1=ipl_bat.groupby("team1").get_group(i).count().id
    t2=ipl_bat.groupby("team2").get_group(i).count().id
    T=t1+t2
    print(f"percentage of winning of {i}: ",(100*w)/T)
```

```python
team_df=pd.DataFrame();
for i in ipl_ball.id.unique():
    team=ipl_ball[ipl_ball["id"]==i]
    
    for j in team.batting_team.unique():
        team1=team[team["batting_team"].str.contains(f"{j}")];
        team_df= team_df.append({"id":i,
                             "batting_team":team1.batting_team.unique(),
                             "batsman_runs":team1.batsman_runs.sum(),
                             "extra_runs":team1.extra_runs.sum(),
                             "total_runs":team1.total_runs.sum()},ignore_index=True);
```
