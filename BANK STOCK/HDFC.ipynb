# 4. Plot the open ammount with Date field after grouping month and yearly .

```python
HDFCBANK.head(5)
```

```python
plt.figure(figsize=(20,5))
plt.plot(HDFCBANK.Date,HDFCBANK.Open)

plt.grid(color = 'red', linestyle = '--', linewidth = 0.5)
plt.xlabel("Date",fontdict = {"size":20})
plt.ylabel("Opening ammount" ,fontdict = {"size":20})
```

```python
plt.figure(figsize=(20,5))
plt.scatter(HDFCBANK.year,HDFCBANK.Open)
plt.xticks(range(2000,2022));
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.xlabel("year",fontdict = {"size":20})
plt.ylabel("Opening ammount" ,fontdict = {"size":20})
```

### In scatter graph we see the opening ammount is very large in 2010 to 2011 as compare to 2000.But the opening ammount in the staring 2011 was decrease in large amount.

```python
plt.figure(figsize=(20,5))
plt.scatter(HDFCBANK.month,HDFCBANK.Open)
plt.xticks(range(0,13));
plt.grid(color = 'red', linestyle = '--', linewidth = 0.5)
plt.xlabel("Month",fontdict = {"size":20})
plt.ylabel("Opening ammount" ,fontdict = {"size":20})
```

### BY seeing this scatter graph we can easly says month 9,10,11 is opening ammount is high as commper to other months.
