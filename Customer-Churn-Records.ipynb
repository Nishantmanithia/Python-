```python
#Select rows where "Age" is between 30 and 40, and the "CreditScore" is greater than 700.
df[(df.Age>30) & (df.Age<40) & (df.CreditScore>700)]
```

```python
#Display the rows for customers who have exited ("Exited" = 1) and have a "Balance" greater than $5,000.
df[(df.Exited==1)&(df.Balance>5000)]
```

```python
#Create a new column "AgeGroup" based on age, categorizing customers as "Young", "Adult", and "Senior".
df["AgeGroup"]=pd.cut(df.Age,bins =[17,40,60,100],labels=["young","adult","Senior"])
df
```

```python
def cat(row):
    if row['Age']<=18 :
         category ="Young"
    elif row['Age'] <=40:
         category= "Adult"
    else:
         category= "Senior"
    return category
df['AgeGroup'] = df.apply(cat , axis=1)
```

```python
df["AgGroup1"]=df.Age.apply(lambda x : "Young" if x<40 else "Adult" if x<60 else "Senior")
df
```

```python
#Calculate the ratio of "Balance" to "EstimatedSalary" for each customer.
df["Bal_Est_ratio"]= df.Balance/df.EstimatedSalary
df
```

```python
#Sort the DataFrame by "CreditScore" in descending order.
df.sort_values(by=["CreditScore"],ascending=False)
```

```python
#Add a new column "BalanceStatus" with values "High" for balances above $10,000 and "Low" otherwise.
df["BalanceStatus"]=df.Balance.apply(lambda x: "Low" if x<10000 else "High")
df
```
