```python
def BMI(x):
    if(x<=18):
        return "underweight"
    elif(x>18 and x<=25):
        return "normal weight"
    elif(x>25 and x<30):
        return "over weight"
    else:
        return "obbese"
```

```python
df['bmi Category']=df.bmi.apply(BMI)
```

```python
df
```

```python
a=[df.groupby('bmi Category').get_group("obbese")['charges'].mean(),
df.groupby('bmi Category').get_group("over weight")['charges'].mean(),
df.groupby('bmi Category').get_group("normal weight")['charges'].mean(),
df.groupby('bmi Category').get_group("underweight")['charges'].mean()]
```
