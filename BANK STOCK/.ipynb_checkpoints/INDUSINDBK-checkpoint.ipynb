```python
# 2. Extract the data from jan 2019 to apr 2021 and plot it.
```

```python
INDUSINDBK.head(5)
```

```python
I1=INDUSINDBK[INDUSINDBK.year>2018]
```

```python
I1.head(5)
```

```python
plt.figure(figsize=(20,5))
plt.plot(I1.Date,I1.Open)
plt.xlabel("Date",fontdict = {"size":20});
plt.ylabel("Opening ammount" ,fontdict = {"size":20});
```

```python
plt.figure(figsize=(20,5))
plt.plot(I1.Date,I1.Volume,"r");
plt.xlabel("Date",fontdict = {"size":20});
plt.ylabel("Volume" ,fontdict = {"size":20});
```

```python
plt.figure(figsize=(20,5))
plt.plot(I1.Date,I1.Trades,"g");
plt.xlabel("Date",fontdict = {"size":20});
plt.ylabel("Trades" ,fontdict = {"size":20});
```

```python
# In above graph we easly can see that Trades and Volume show a large increment change and the value of opening ammount start 
drop in the start of 2020 due to the pandemic covid-19.
```

```python
# 3. plot the open ammount with Date field after grouping month,yearly and weekly.
```

```python
INDUSINDBK.head(5)
```

```python
plt.figure(figsize=(20,5))
plt.scatter(INDUSINDBK.year,INDUSINDBK.Open)
plt.xticks(range(2000,2022));
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.xlabel("year",fontdict = {"size":20})
plt.ylabel("Opening ammount" ,fontdict = {"size":20})
```

```python
# In scatter graph we see the opening ammount is very large in 2001 to 2018.But the opening ammount in the 
staring 2019 was decrease in large amount due to covid crisis.
```

```python
plt.figure(figsize=(20,5))
plt.scatter(INDUSINDBK.month,INDUSINDBK.Open)
plt.xticks(range(0,13));
plt.grid(color = 'red', linestyle = '--', linewidth = 0.5)
plt.xlabel("Month",fontdict = {"size":20})
plt.ylabel("Opening ammount" ,fontdict = {"size":20})
```

```python
# BY seeing this scatter graph we can easly says month 6,7,8 is opening ammount is high as commper to other months.
```
