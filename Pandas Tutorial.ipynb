# Series

It is a one-dimensional array holding data of any type.

```python
import pandas as pd
x = [1, 8, 2]
var = pd.Series(x) 
print(var)
```

# Create Labels

```python
import pandas as pd
x = [1, 8, 2]
var = pd.Series(x, index = ["x", "y", "z"])
print(var)
```

# DataFrames

Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

```python
import pandas as pd
dt = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
var = pd.DataFrame(dt)
print(var)
```

# Locate Row

```python
print(var.loc[1])
```

```python
print(var.loc[[0,2,1]])
```
